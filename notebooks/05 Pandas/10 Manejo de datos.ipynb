# Manejo de datos

Dataset: https://www.kaggle.com/berkeleyearth/climate-change-earth-surface-temperature-data?select=GlobalLandTemperaturesByCountry.csv

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/GlobalLandTemperaturesByCountry.csv')
df
```

```python
# Tamañao del dataset
df.shape
```

```python
# Conocer las variables del dataset
df.info()
```

```python
df['Country'].unique()
```

```python
df['year'] = pd.to_datetime(df['dt'])
df['year']
```

```python
idx = df['year'] > pd.to_datetime('1970-01-01')
df = df[idx]
df
```

```python
df_t_avg = df.groupby(['Country', pd.Grouper(key='year', freq='1Y')]).agg({'AverageTemperature': [np.mean, np.median]})
df_t_avg
```

```python
df_t_avg.xs('Mexico')['AverageTemperature'].plot()
```

```python
df_t_med = df_t_avg['AverageTemperature'][['median']].reset_index()
df_t_med
```

```python
df_t_med['date'] = df_t_med['year'].dt.year
df_t_med['date']
```

```python
df_t_med.rename(columns={'median': 'temperature'}, inplace=True)
df_t_med
```

```python
df_t_pivot = df_t_med.pivot_table(values='temperature', index='date', columns='Country')
df_t_pivot
```

```python
df_t_pivot.boxplot()
```
