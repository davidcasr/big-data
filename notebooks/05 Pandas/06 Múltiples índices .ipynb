# Multiples indices

```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('data/poblacion.csv')
df
```

```python
df['year'] = pd.Categorical(df['year'].apply(str))
```

```python
df.dtypes
```

```python
filtro = df['Country'].isin(['Aruba', 'Colombia'])
filtro
```

```python
sample = df[filtro]
sample
```

```python
sample = sample.set_index(['Country', 'year']).sort_index()
sample
```

```python
sample.loc['Colombia', :]
```

```python
sample.loc['Colombia', :].loc['2016', :]
```

```python
sample.xs(['Aruba'])
```

```python
sample.xs('2018', level='year')
```

```python
countries = df.set_index(['Country', 'year']).sort_index()
countries
```

```python
ids = pd.IndexSlice
countries.loc[ids['Aruba':'Austria', '2015':'2017'], :].sort_index()
```

```python
countries.sum(level='year')
```
