# GroupBy y Aggregation 

```python
import pandas as pd 
import numpy as np
```

```python
import seaborn as sns
```

## Dataset: Exercise

```python
df = sns.load_dataset('exercise')
df
```

```python
sns.get_dataset_names()
```

```python
df.groupby('diet').mean()
```

```python
df.groupby('diet').median()
```

```python
for key_group, group in df.groupby('diet'):
    pulse = group['pulse'].mean()
    print('Diet: {}, Pulse: {} \n'.format(key_group, pulse))
```
